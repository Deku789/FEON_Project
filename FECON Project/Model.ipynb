```python
final_dataset = pd.merge(MIBORDataframe[['Date','3-Month-MIBOR']],treasury_3MonthRate,on='Date',how='outer')
final_dataset = pd.merge(final_dataset,Gbond10YearYield,on='Date',how='outer')
final_dataset = pd.merge(final_dataset,NiftyBank[['Date','Open_bank','Close_bank']],on='Date',how='outer')
final_dataset = pd.merge(final_dataset,Nifty500[['Date','Open_500','Close_500']],on='Date',how='outer')
final_dataset.sort_values(by="Date",inplace=True)
final_dataset = final_dataset[final_dataset['Date']> "2000-01-01"]
```

```python
final_dataset[['Return_Bank','Return_500']].cov().iloc[0,1]/final_dataset[['Return_Bank']].var()
```

```python
final_dataset['Beta_bank'] = 0.532567

final_dataset['3-Month Treasury Bill Rate'] = final_dataset['3-Month Treasury Bill Rate'].fillna(final_dataset['3-Month Treasury Bill Rate'].mean())
final_dataset['3-Month-MIBOR'] = final_dataset['3-Month-MIBOR'].fillna(final_dataset['3-Month-MIBOR'].mean())
final_dataset['10 Year Gbond Yield'] = final_dataset['10 Year Gbond Yield'].fillna(final_dataset['10 Year Gbond Yield'].mean())
final_dataset['Open_bank'] = final_dataset['Open_bank'].fillna(final_dataset['Open_bank'].mean())
final_dataset['Close_bank'] = final_dataset['Close_bank'].fillna(final_dataset['Close_bank'].mean())
final_dataset['Open_500'] = final_dataset['Open_500'].fillna(final_dataset['Open_500'].mean())
final_dataset['Close_500'] = final_dataset['Close_500'].fillna(final_dataset['Close_500'].mean())
final_dataset['Return_Bank'] = (final_dataset['Close_bank'] - final_dataset['Open_bank'])/final_dataset['Open_bank']
final_dataset['Return_500'] = (final_dataset['Close_500'] - final_dataset['Open_500'])/final_dataset['Open_500']
final_dataset[final_dataset['Date']> "2000-01-01"]
```

```python
final_dataset['c_max_bank'] = final_dataset['Return_Bank']/final_dataset['Return_Bank'].max()
final_dataset['c_max_500'] = final_dataset['Return_500']/final_dataset['Return_500'].max()
final_dataset
```

```python
final_dataset['c_max_final'] = (final_dataset['c_max_bank'] - final_dataset['c_max_bank'].mean())/final_dataset['c_max_bank'].std()
final_dataset['c_max_final_500'] = (final_dataset['c_max_500'] - final_dataset['c_max_500'].mean())/final_dataset['c_max_500'].std()
final_dataset['rvol_bank'] = final_dataset['Return_Bank'].rolling(min_periods=1, window=30).sum()
final_dataset['rvol_500'] = final_dataset['Return_500'].rolling(min_periods=1, window=30).sum()
final_dataset['rvol_bank'] = (final_dataset['rvol_bank'] - final_dataset['rvol_bank'].mean())/final_dataset['rvol_bank'].std()
final_dataset['rvol_500'] = (final_dataset['rvol_500'] - final_dataset['rvol_500'].mean())/final_dataset['rvol_500'].std()
final_dataset
```
